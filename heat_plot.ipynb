```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('contact-probability.dat',sep='\s+')
```

```python
cont=df.pivot('x','y','p2')
plt.subplots(figsize=(10,10))
display=sns.heatmap(cont,cmap='crest',vmin=0.0,vmax=1.0,cbar=True)
```

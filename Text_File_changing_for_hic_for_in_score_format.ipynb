```python
# Read the input file
with open('chr_19_mod.txt', 'r') as file:
    lines = file.readlines()

# Initialize an empty list to store the converted data
result = []

# Extract column headers (x, y, z) from the first line
columns = lines[0].split()[1:]

# Iterate through the data and convert it
for line in lines[1:]:
    parts = line.split()
    row_label = parts[0]
    
    for col_label, value in zip(columns, parts[1:]):
        # Keep the first two columns as strings and convert the third column to a float
        if col_label == columns[0]:
            formatted_value = row_label
        elif col_label == columns[1]:
            formatted_value = value
        else:
            formatted_value = f'{float(value):.5f}'
        result.append(f'{row_label} {col_label} {formatted_value}')

# Write the result to an output file or print it
with open('output_file.txt', 'w') as file:
    for item in result:
        file.write("%s\n" % item)
```
